# Nye zero-stress crevasse depth criteria in a coupled flow-flexure system (viscous)

The deviatoric stress field for a coupled SSA flow-flexure system is 

$$\tau = 2\overline{\nu}\varepsilon - 2\nu_f\nabla^2\dot{\eta}\zeta,$$

where $\overline{\nu}$ is the depth-averaged effective viscosity associated with flow, $\varepsilon$ is the membrane strain rate tensor, $\nu_f$ is the flexural viscosity, $\dot{\eta}$ is the vertical deflection rate, and $\zeta$ is a vertical coordinate which ranges from $-\frac{H}{2}$ to $\frac{H}{2}$ ($H$ the shelf thickness). Under the familiar assumions used throughout the notebooks in this repository, the corresponding pressure is given as 

$$P = -tr(2\overline{\nu}\varepsilon) + tr(2\nu_f\nabla^2\dot{\eta})\zeta + \rho_ig\left(\frac{H}{2} - \zeta\right).$$

The Nye zero-stress crevasse depth criterion states that crevasses penetrate to the depth at which the principal tension (the first eigenvalue of $\tau$) is equal to the pressure. In this notebook, to permit pseudo-analytic solutions, I'll treat crevasses from flow as separate from crevasses from flexure, so that $\tau = 2\overline{\nu}\varepsilon$ in the context of flow, and $\tau = -2\nu_f\nabla^2\dot{\eta}\zeta$ in the context of flexure. However, pressure will remain a function of both the flow field and the flexure field. Introducing the shorthand notation $\tau_f := 2\overline{\nu}\varepsilon$ (the flow contribution to the stress tensor) and $\tau_\nu^* := -2\nu_f\nabla^2\dot{\eta}$ (i.e., the $\zeta$-independent part of the flexural contribution), and using prime notation to indicate the largest eigenvalue of a tensor, our crevasse depth law will state that crevasses from flow reach the vertical coordinate $\zeta_c$ at which

$$\tau_f' = -tr(\tau_f) - tr(\tau_\nu^*)\zeta_c + \rho_ig\left(\frac{H}{2} - \zeta_c\right),$$

while crevasses from flexure reach the vertical coordinate at which 

$$(\tau_\nu^*\zeta_c)' = -tr(\tau_f) - tr(\tau_\nu^*)\zeta_c + \rho_ig\left(\frac{H}{2} - \zeta_c\right).$$

Beginning with these crevasse depth laws, I'll derive expressions for surface crevasse depths from flexure, basal crevasse depths from flexure, surface crevasse depths from flow, and basal crevasse depths from flow. But first we need to look at the conditions under which this form of crevasse law makes sense:

## Caveats

Notice that the question "How deep can a crevasse reach before it encounters pressure sufficient to halt it?" is not automatically well-posed. For a surface crevasse, the question assumes that pressure increases with depth; i.e., $\frac{\partial}{\partial \zeta}P < 0$ (note that $\zeta$ is elevation, rather than depth). However, examining the equations above, this is only the case when $-tr(\tau_\nu^*) < \rho_ig$. That is, pressure can remain constant or even decrease with depth for sufficient upward flexure. Since $\tau_f'$ is depth-invariant (assuming an isothermal shelf), there will be no positive, finite solution for flow-driven surface crevasse depth unless pressure increases downward. Therefore, the flow field will produce negative or infinite crevasse depth whenever both of the following conditions are true together: 

1) $P|_{surf} < \tau_f'$ (i.e., conditions at the surface are conducive to flow-driven crevasse formation), and 
2) $tr(\tau_\nu^*) + \rho_ig \leq 0$ (i.e., pressure does not increase with depth).

The discussion is slightly different for flexure-driven crevasse formation: since $(\tau_\nu^*\zeta)'$ decreases with depth, we do not necessarily need $P$ to increase with depth in order to have positive crevasse depths from the flexure field: we only need $P$ not to decrease faster than $(\tau_\nu^*\zeta)'$. That is, crevasses will be well-behaved as long as $\frac{\partial}{\partial\zeta}P < \frac{\partial}{\partial\zeta}(\tau_\nu^*\zeta)'$. To see when this will be the case, let $\lambda_1 < \lambda_2$ represent the first and second eigenvalues of $\tau_\nu^*$. Then $\tau_\nu^*\zeta$ has eigenvalues $\lambda_1\zeta$ and $\lambda_2\zeta$, with the larger of the two determined by the sign of $\zeta$. Without loss of generality, let's work in the top half of the shelf, so that $\zeta > 0$ and $\lambda_1\zeta$ is the first eigenvalue of $\tau_\nu^*\zeta$. Then $\frac{\partial}{\partial\zeta}(\tau_\nu^*\zeta)' = \lambda_1$, and the condition for well-behaved crevasses is that $\frac{\partial}{\partial\zeta}P < \lambda_1$. Therefore, the flexure field will produce unphysical surface crevasse depths whenever both of the following conditions are true together:

1) $P|_{surf} < \lambda_1\frac{H}{2}$ (i.e., the maximum tension from flexure at the surface is greater than the pressure there), and
2) $\lambda_{1,2} + tr(\tau_\nu^*) + \rho_ig \leq 0$ (i.e., pressure does not decrease faster with depth than the maximum tension does).

In these cases, it is probably most appropriate to assume that surface crevasses simply reach the base (and a similar discussion holds for basal crevasses). These caveats will be used as conditionals on the solutions below.

## Surface crevasses from flexure

Define the crevasse depth $d$ so that $\zeta_c = \frac{H}{2} - d$. Suppose that the principal tension from flexure at the surface is greater than the pressure, so that $P|_{surf} < \lambda_1\frac{H}{2}$ (i.e., conditions at the surface are conducive to crevassing). Expressing the flexural crevasse law in terms of $d$, and bearing in mind the caveats discussed above, we will have crevasses reaching the depth at which

$$(\tau_\nu^*\zeta_c)' = -tr(\tau_f) -tr(\tau_\nu^*)\frac{H}{2} + \left(tr(\tau_\nu^*) + \rho_ig\right)d.$$

To evaluate the left-hand side of this equality, consider two cases: either the crevasse tip lies above the vertical midpoint of the shelf, or the crevasse tip lies below the vertical midpoint of the shelf. If it is the former, then $\zeta_c > 0$; since $\lambda_1 > \lambda_2$ by definition, we may conclude that $\lambda_1\zeta_c > \lambda_2\zeta_c$. This guarantees that the principal tension at the crevasse tip is $(\tau_\nu^*\zeta_c)' = \lambda_1\zeta_c = \lambda_1\left(\frac{H}{2} - d\right)$. Conversely, if the crevasse tip lies below the shelf's midpoint, then $\zeta_c < 0$, and the principal tension at the crevasse tip is $(\tau_\nu^*\zeta_c)' = \lambda_2\zeta_c = \lambda_2\left(\frac{H}{2} - d\right)$. Thus, we find that 

$$d = \begin{cases}
\frac{\left(\frac{1}{2}\lambda_1 + \frac{1}{2}tr(\tau_\nu^*)\right)H + tr(\tau_f)}{\lambda_1 + tr(\tau_\nu^*) + \rho_ig} & d \in \left[0, \frac{H}{2}\right], P|_{surf} < \lambda_1\frac{H}{2},\text{ and } \lambda_1 + tr(\tau_\nu^*) + \rho_ig > 0\\
\\
\frac{\left(\frac{1}{2}\lambda_2 + \frac{1}{2}tr(\tau_\nu^*)\right)H + tr(\tau_f)}{\lambda_2 + tr(\tau_\nu^*) + \rho_ig} & d \in \left[\frac{H}{2}, H\right], P|_{surf} < \lambda_1\frac{H}{2}, \text{ and } \lambda_2 + tr(\tau_\nu^*) + \rho_ig > 0\\
\\
0 & P_{surf} \geq \lambda_1\frac{H}{2}\\
\\
H & P_{surf} < \lambda_1\frac{H}{2} \text{ and } \lambda_{1,2} + tr(\tau_\nu^*) + \rho_ig \leq 0
\end{cases}$$

Notice that, if not for the conditionals, the solution would yield negative crevasse depths when $\lambda_{1,2} + tr(\tau_\nu^*) + \rho_ig < 0$, and there would be a singularity when $\lambda_{1,2} + tr(\tau_\nu^*) + \rho_ig = 0$ (which occurs when the pressure and the principal flexural tension change at the same rate with depth). 

## Basal crevasses from flexure

Suppose instead that the principal tension from flexure at the base is greater than the basal pressure, so that conditions at the base are conducive to crevassing. In this case, this occurs when $P|_{base} < -\lambda_2\frac{H}{2}$. We'll redefine the crevasse depth $d$ to increase upward, so that the vertical coordinate at the crevasse tip is $\zeta_c = d - \frac{H}{2}$ (e.g., if a crevasse were to reach upward halfway toward the shelf midpoint, so that $\zeta_c = -\frac{H}{4}$, this crevasse would have depth $\frac{1}{4}H$, or one-quarter the shelf thickness). The zero-stress criterion must then be reframed so that crevasses penetrate upward to the depth at which 

$$(\tau_\nu^*\zeta_c)' = -tr(\tau_f) + \left(\frac{1}{2}tr(\tau_\nu^*) + \rho_ig\right)H - \left(tr(\tau_\nu^*) + \rho_ig\right)d.$$

As above, our expression for the principal tension $(\tau_\nu^*\zeta_c)$ will depend on which half of the shelf the crevasse tip lies in. If $\zeta_c > 0$, then $(\tau_\nu^*\zeta_c)' = \lambda_1\zeta_c = \lambda_1\left(d - \frac{H}{2}\right).$ If $\zeta_c < 0$, then $(\tau_\nu^*\zeta_c)' = \lambda_2\zeta_c = \lambda_2\left(d - \frac{H}{2}\right).$ Therefore, a basal crevasse due to flexure will reach the depth (measured upward from the base) at which 

$$d = \begin{cases}
\frac{\left(\frac{1}{2}\lambda_2 + \frac{1}{2}tr(\tau_\nu^*) + \rho_ig\right)H -tr(\tau_f)}{\lambda_2 + tr(\tau_\nu^*) + \rho_ig} & d \in \left[0, \frac{H}{2}\right], P|_{base} < -\lambda_2\frac{H}{2}, \text{ and } \lambda_2 + tr(\tau_\nu^*) + \rho_ig < 0\\
\\
\frac{\left(\frac{1}{2}\lambda_1 + \frac{1}{2}tr(\tau_\nu^*) + \rho_ig\right)H -tr(\tau_f)}{\lambda_1 + tr(\tau_\nu^*) + \rho_ig} & d \in \left[\frac{H}{2}, H\right], P|_{base} < -\lambda_2\frac{H}{2}, \text{ and } \lambda_1 + tr(\tau_\nu^*) + \rho_ig < 0\\
\\
0 & P|_{base} \geq -\lambda_2\frac{H}{2}\\
\\
H & P|_{base} < -\lambda_2\frac{H}{2} \text{ and } \lambda_{1,2} + tr(\tau_\nu^*) + \rho_ig \geq 0
\end{cases}
$$

## Surface crevasses from flow

With $d$ once more representing the downward extent of a crevasse, a surface crevasse from flow will stop propagating when 

$$\tau_f' = -tr(\tau_f) -tr(\tau_\nu^*)\frac{H}{2} + \left(tr(\tau_\nu^*) + \rho_ig\right)d.$$

Bearing in mind the caveats discussed earlier, our surface crevasse depth law states that 

$$d = \begin{cases}
\frac{\tau_f' + tr(\tau_f) + tr(\tau_\nu^*)\frac{H}{2}}{tr(\tau_\nu^*) + \rho_ig} & P|_{surf} \leq \tau_f', \text{ and } tr(\tau_\nu^*) + \rho_ig > 0\\
\\
0 & P|_{surf} \geq \tau_f'\\
\\
H & P|_{surf} < \tau_f' \text{ and } tr(\tau_\nu^*) + \rho_ig \leq 0.
\end{cases}
$$

In the case of 1D, non-flexural flow, $\tau_f' = tr(\tau_f) = \tau_{xx}$ and $\tau_\nu^* = 0$, and this crevasse depth law reduces to the familiar statement that $d = \frac{2\tau_{xx}}{\rho_ig}$. 

## Basal crevasses from flow

Revert back to our basal definition of $d$, with $d = \zeta_c + \frac{H}{2}$. Then a crevasse due to flow should reach upward to the point at which 

$$\tau_f' = -tr(\tau_f) + \left(\frac{1}{2}tr(\tau_\nu^*) + \rho_ig\right)H - \left(tr(\tau_\nu^*) + \rho_ig\right)d.$$

This gives rise to a basal crevasse depth law stating that  

$$d = \begin{cases}
\frac{-\tau_f' - tr(\tau_f) + \left(\frac{1}{2}tr(\tau_\nu^*) + \rho_ig\right)H}{tr(\tau_\nu^*) + \rho_ig} & P|_{base} < \tau_f', \text{ and } tr(\tau_\nu^*) + \rho_ig < 0\\
\\
0 & P|_{base} \geq \tau_f'\\
\\
H & P|_{base} < \tau_f' \text{ and } tr(\tau_\nu^*) + \rho_ig \geq 0.
\end{cases}
$$
